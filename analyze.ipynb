```python
print ("A patch is labelled with the class that has maximum presence in the corresponding ground-truth patch")
```

```python
print ("For example, the corresponding ground truth of the patch above is:")
print(pd.gt_patches[20])
```

```python
dominant = pd.gt_patches[20].reshape(-1).bincount().argmax()
print (f"For example, the class {dominant} has maximum presence in the corresponding ground truth")
print(f"Therefore, we label the patch above as class {dominant}")
```
